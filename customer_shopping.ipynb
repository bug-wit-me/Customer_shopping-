```python
purchase_group_map = {
    'Weekly': 'Regular Purchases',
    'Fortnightly': 'Regular Purchases',
    'Bi-Weekly': 'Regular Purchases',
    'Monthly': 'Occasional Purchases',
    'Quarterly': 'Occasional Purchases',
    'Every 3 Months': 'Occasional Purchases',
    'Annually': 'Rare Purchases'
}

df['Frequency Group'] = df['Frequency of Purchases'].map(purchase_group_map)
```

```python
def age_group(age):
  if age<=18:
    return "Teenager"
  elif age>18 and age<=30:
    return "Young"
  elif age>30 and age<=50:
    return "Middle"
  else:
    return "Old"
```

```python
df["Age Group"]=df["Age"].apply(age_group)
df
```

```python
grouped_data = df.groupby(['Age Group', 'Gender'])['Frequency of Purchases'].size().reset_index(name='Count')
pivot_data = grouped_data.pivot(index='Age Group', columns='Gender', values='Count').fillna(0)
pivot_data.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')
plt.title('Frequency of Purchases by Age Group and Gender')
plt.xlabel('Age Group')
plt.ylabel('Count of Purchases')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

```python
grouped_data = df.groupby(['Age Group', 'Gender', 'Frequency Group']).size().reset_index(name='Count')
pivot_data = grouped_data.pivot_table(index='Age Group', columns=['Gender', 'Frequency Group'], values='Count', fill_value=0)
pivot_data.plot(kind='bar', stacked=True, figsize=(16, 7), colormap='viridis')
plt.title('Frequency of Purchases by Age Group & Gender')
plt.xlabel('Age Group')
plt.ylabel('Count of Purchases')
plt.legend(title='Gender and Frequency of Purchases', bbox_to_anchor=(1.05, 1), loc='best')
plt.tight_layout()
plt.show()
```
